## Solução

### Questão 1 a)

Utilizando $P_{atm} = 1\,\mathrm{atm} = 101.325\,\mathrm{Pa}$

$$
H_{atm} = \frac{P_{atm}}{\rho g} = 12,01 \,\mathrm{m}
$$

```python
Patm = 101325  # Pa
g = 9.81       # m/s^2
rho = 860      # kg/m^3

Hatm = Patm / (rho * g)

print(f'Hatm = {Hatm:.3f} m')
```

### Questão 1 b)

$$
V_{cil} = \left(\frac{\pi D_{cil}^2}{4}\right)L = 9,425\,\mathrm{l}
$$

$$
\bar{Q} = V_{cil} f = 0,785 \,\mathrm{l/s}
$$

```python
from numpy import pi

Dcil = 0.2  # m
L = 0.3     # m

Acil = pi * Dcil**2 / 4
Vcil = Acil * L

print(f'Vcil = {1000*Vcil:.3f} l')

rot = 5  # rpm
f = rot / 60  # Hz

Qm = Vcil * f

print(f'Qm = {1000*Qm:.3f} l/s')
```

### Questão 1 c)

Altura estática de elevação: $h_d = 27\,\mathrm{m}$

Delta de aceleração para descarga:
$$
H_{acc,d} = \left(\frac{A_{cil}}{A_{tubo,d}}\right)\frac{L_d}{g} r\omega^2 = 2,146 \,\mathrm{m}
$$

Altura manométrica no início da descarga:

$$
H_{d,i} = H_{atm} + h_d + H_{acc,d} = 41,156 \,\mathrm{m}
$$


```python
Dtubo = 0.05  # m
Atubo = (pi*Dtubo**2)/4

hd = 27     # m
Ld = 32     # m
r = L/2     # O raio da manivela é metade do curso do cilindro
w = 2*pi*f  # rad/s

Hacc_d = (Acil/Atubo) * (Ld/g) * r * w**2

print(f'Hacc_d = {Hacc_d:.3f} m')

Hd_i = Hatm + hd + Hacc_d

print(f'Altura manométrica no início da descarga = {Hd_i:.3f} m')
```

### Questão 1 d)

Altura manométrica no final da descarga:

$$
H_{d,f} = H_{atm} + h_d - H_{acc,d} = 36,864 \,\mathrm{m}
$$

```python
Hd_f = Hatm + hd - Hacc_d

print(f'Altura manométrica no final da descarga = {Hd_f:.3f} m')
```

### Questão 1 e)

Delta de aceleração para a sucção:
$$
H_{acc,s} = \left(\frac{A_{cil}}{A_{tubo,s}}\right)\frac{L_s}{g} r\omega^2 =  0,335 \,\mathrm{m}
$$

Altura manométrica no início da descarga:

$$
H_{s,i} = H_{atm} - h_s - H_{acc,s} =  8,675 \,\mathrm{m}
$$


```python
hs = 3     # m
Ls = 5     # m

Hacc_s = (Acil/Atubo) * (Ls/g) * r * w**2

print(f'Hacc_s = {Hacc_s:.3f} m')

Hs_i = Hatm - hs - Hacc_s

print(f'Altura manométrica no início da sucção = {Hs_i:.3f} m')
```

### Questão 1 f)

Altura manométrica no final da sucção:

$$
H_{s,f} = H_{atm} - h_s + H_{acc,s} =  9,346 \,\mathrm{m}
$$

```python
Hs_f = Hatm - hs + Hacc_s

print(f'Altura manométrica no final da sucção = {Hs_f:.3f} m')
```

### Questão 1 g)

Cálculo do número de Reynolds na tubulação:

$$
V = \frac{\bar{Q}}{A_{cil}} = 0,025\,\mathrm{m/s}
$$

$\phantom{a}$

$$
\mu = \nu \rho = 70\times 10^{-6} \,\mathrm{m^2/s} \times 860\,\mathrm{kg/m^3} = 0,0602\,\mathrm{Pa.s}
$$

$\phantom{a}$

$$
\mathrm{Re} = \frac{\rho V D}{\mu} = 71,429
$$

Como $\mathrm{Re}<2300$, o fator de atrito é dado por:
$$
f_f = \frac{64}{\mathrm{Re}} = 0,896
$$

Delta de perda de carga na descarga:
$$
H_{T,d} = f_f \left(\frac{L_d}{D_{tubo}}\right)\left(\frac{A_{cil}}{A_{tubo,d}}\right)^2\frac{r^2\omega^2}{2g}  =  46,154 \,\mathrm{m}
$$

Altura manométrica no centro da descarga:
$$
H_{ld} = H_{d,c} = H_{atm} + h_d + H_{T,d} = 85.164 \,\mathrm{m}
$$

```python
V = Qm/Acil    # m/s
u = 70e-6*860  # Pa.s

Re = rho * V * Dcil / u

print(f'Re = {Re:.3f}')

ff = 64/Re

print(f'f_f = {ff:.3f}')

Ht_d = ff * (Ld/Dtubo) * (Acil/Atubo)**2 * r**2 * w**2 / (2*g)

print(f'Ht_d = {Ht_d:.3f} m')

Hd_c = Hatm + hd + Ht_d

print(f'Altura manométrica no centro da descarga = {Hd_c:.3f} m')
```

### Questão 1 h)

Delta de perda de carga no centro da sucção:

$$
H_{T,s} = f_f \left(\frac{L_s}{D_{tubo}}\right)\left(\frac{A_{cil}}{A_{tubo,s}}\right)^2\frac{r^2\omega^2}{2g}  =  7,212 \,\mathrm{m}
$$

Altura manométrica no centro da sucção:

$$
H_{ls} = H_{s,c} = H_{atm} - h_s - H_{T,s} = 1,799 \,\mathrm{m}
$$

```python
Ht_s = ff * (Ls/Dtubo) * (Acil/Atubo)**2 * r**2 * w**2 / (2*g)

print(f'Ht_s = {Ht_s:.3f} m')

Hs_c = Hatm - hs - Ht_s

print(f'Altura manométrica no centro da sucção = {Hs_c:.3f} m')
```

### Questão 1 j)

Há três pontos críticos:
- Final da descarga
- Início da sucção
- Centro da sucção

---
#### **Final da descarga**

No final da descarga, a altura manométrica em função da rotação $\omega$ é dada por:

$$
H_{d,f} = H_{atm} + h_d - H_{acc,d}
$$

Portanto:

$$
H_{d,f} = H_{atm} + h_d - \left(\frac{A_{cil}}{A_{tubo,d}}\right)\frac{L_d}{g} r\omega^2 > H_{lim}
$$

Sendo:

$$
H_{lim} = \frac{20.000\, \mathrm{Pa}}{860 \frac{kg}{m^3}. 9,81\frac{m}{s^2}} = 2,371\,\mathrm{m}
$$

Isolando $\omega$, obtem-se:

$$
\omega < \sqrt{\frac{H_{atm} + h_d -H_{lim}}{\displaystyle \left(\frac{A_{cil}}{A_{tubo,d}}\right)\frac{L_d}{g} r}}
$$

Assim:

$$
\omega < 20,659\,\mathrm{rpm}
$$


```python
from numpy import sqrt

Hlim = 20000/(rho*g)

print(f'Hlim = {Hlim:.3f} m')

w1 = sqrt((Hatm + hd - Hlim) / ((Acil/Atubo) * (Ld/g) * r))

print(f'w1 < {w1:.3f} rad/s')
print(f'w1 < {w1/2/pi:.3f} Hz')
print(f'w1 < {60*w1/2/pi:.3f} rpm')
```

---

#### **Início da Sucção**

No ínício da sucção, a pressão interna ao cilindro em altura de coluna é dada por:

$$
H_{s,i} = H_{atm} - h_s - H_{acc,s} > H_{lim}
$$

Fazendo as mesmas manipulações, chega-se a:

$$
\omega < \sqrt{\frac{H_{atm} - h_s -H_{lim}}{\displaystyle \left(\frac{A_{cil}}{A_{tubo,s}}\right)\frac{L_s}{g} r}}
$$

O que resulta em:

$$
\omega< 22,248 \,\mathrm{rpm}
$$

```python
w2 = sqrt((Hatm - hs - Hlim) / ((Acil/Atubo) * (Ls/g) * r))

print(f'w2 < {w2:.3f} rad/s')
print(f'w2 < {w2/2/pi:.3f} Hz')
print(f'w2 < {60*w2/2/pi:.3f} rpm')
```

---
#### **Centro da Sucção**

No centro da sucção, a pressão interna ao cilindro em altura manométrica é dada por:

$$
H_{s,c} = H_{atm} - h_s - H_{T,s} = H_{atm} - h_s - f_f \left(\frac{L_s}{D_{tubo}}\right)\left(\frac{A_{cil}}{A_{tubo,s}}\right)^2\frac{r^2\omega^2}{2g} > H_{lim}
$$

Portanto:

$$
\omega < \sqrt{\frac{H_{atm} - h_s -H_{lim}}{\displaystyle f_f \left(\frac{L_s}{D_{tubo}}\right)\left(\frac{A_{cil}}{A_{tubo,s}}\right)^2\frac{r^2}{2g}}}
$$

O que resulta em:

$$
\omega < 4,798\,\mathrm{rpm}
$$

```python
w3 = sqrt((Hatm-hs-Hlim)/(ff * (Ls/Dtubo) * (Acil/Atubo)**2 * r**2 / (2*g)))

print(f'w3 < {w3:.3f} rad/s')
print(f'w3 < {w3/2/pi:.3f} Hz')
print(f'w3 < {60*w3/2/pi:.3f} rpm')
```

Assim, a maior vazão admissível será de:

$$
Q_{max} = f_{max} V_{cil} = \frac{4,798\,\mathrm{rpm}}{60 \,\mathrm{rpm/Hz}{}}\cdot 9,425\,\mathrm{l} = 0,7536\,\mathrm{l/s}
$$

```python
Qmax = (w3/2/pi)*Vcil

print(f'Qmax = {1000*Qmax:.4f} l/s')
```
