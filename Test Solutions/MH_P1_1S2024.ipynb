### Questão 1) - Solução

Considere o diagrama a seguir:

image.png

A pressão manométrica no ponto 1 é a pressão $P_A = 3,0\, kPa$, e a pressão no final do manômetro $B$ é a pressão atmosférica. Assim, pelo teorema dos vasos comunicantes, para a gasolina:

$$
P_1 = P_A = \rho_B g h_1 \;\;\;\; \therefore \;\;\;\; h_1 = \frac{P_A}{\rho_B g}
$$

Da geometria do problema:

$$
z_b = h_1 + 1,5\,m + 1,0 \,m
$$

Já no ponto 2 a pressão manométrica é dada pela pressão $P_A$ somada com a pressão gerada pela coluna de gasolina. No manômetro $C$, pelo princípio dos vasos comunicantes, no nível do ponto 2 a pressão manométrica é dada por $\rho_C g h_2$. Assim:

$$
P_2 = P_A + \rho_B g . 1,5\, m = \rho_C g h_2
$$

Então, podemos calcular $h_2$:

$$
h_2 = \frac{P_A + \rho_B g . 1,5\, m}{\rho_C g}
$$

E assim:

$$
z_c = 1,0\, m + h_2
$$

Os resultados seguem abaixo.

```python
# 1)

%reset -f

rho_b = 770   # kg/m^3
rho_c = 1260  # kg/m^3
h_b = 1.5     # m
h_c = 1.0     # m
g = 9.81      # m/s^2
pa = 3000     # Pa

p1 = pa
zb = h_b+h_c+p1/(rho_b*g)

print('zb = ', zb, 'm')

p2 = pa + rho_b*h_b*g
zc = h_c + p2/(rho_c*g)

print('zc = ', zc, 'm')
```

```python
# 2)

%reset -f

# Dados

rho = 998.37    # kg/m^3
vs = 17         # m/s
d1 = 12*0.0254  # m
d2 = 6*0.0254   # m

# Medidas Auxiliares

from numpy import pi

A1 = pi*(d1**2)/4
A2 = pi*(d2**2)/4
```

### Questão 2.a) - Solução

O bocal trabalha em regime com fluxo incompressível. Por isso a vazão volumétrica é a vazão de saída, dada por:

$$
Q = v_s A_2 = v_s \left(\frac{\pi d_2^2}{4}\right)
$$

Os resultados seguem abaixo.

```python
# 2.a)

Q = vs*A2

print('Q = ', Q*1000, 'L/s')
```

### Questão 2.b) - Solução

Como as vazões volumétricas se conservam, dado o fluxo incompressível, e agora conhecemos a vazão $Q$, a velocidade de entrada é dada por:

$$
v_e = \frac{Q}{A_1}
$$

Os resultados seguem abaixo.

```python
# 2.b)

ve = Q/A1

print('ve = ', ve, 'm/s')
```

### Questão 2.c) - Solução

Para calcular a pressão na entrada do bocal, uma vez que a perda de carga pe desprezível, pode-se utilizar a equação de Bernoulli na linha central do bocal entre o ponto de entrada e o ponto de saída do fluido. Sobre esta linha, que é horizontal, temos $z_1 = z_2 = 0$, adotado como referência.

Na entrada temos $p_1 = p_e$ e $v_1 = v_e$. Na saída temos $p_2 = P_{atm}$ e $v_2=v_s$ conhecida. Assim:

$$
\frac{p_1}{\rho} + \frac{v_1^2}{2} + g z_1 = \frac{p_2}{\rho} + \frac{v_2^2}{2} + g z_2
$$

Substituindo os dados:

$$
\frac{p_e}{\rho} + \frac{v_e^2}{2} = \frac{P_{atm}}{\rho} + \frac{v_2^2}{2}
$$

Assim:

$$
p_e - P_{atm} = p_e^{(mano)} = \frac{\rho}{2}(v_2^2-v_e^2)
$$

Os resultados seguem:

```python
# 2.c)

pe = rho*(vs**2-ve**2)/2

print('pe = ', pe, 'Pa (mano)')
```

### Questão 2.d) - Solução

Considere o diagrama a seguir:

image.png

Define-se o volume de controle conforme o diagrama. O fluido flui através da área da entrada do bocal com velocidade $v_e$ e através da área da saída com a velocidade $v_s$. A pressão manométrica $p_e$ age sobre a área de entrada do bocal. A pressão atmosférica age sobre todas as superfícies do volume de controle e seu efeito se anula. A força de reação horizontal dos parafusos (total) é representada por $F_p$.

Com estas definções, aplica-se o princípio da conservação da quantidade de movimento na horizontal:

$$
\sum F_{c, x} + \sum F_{s, x} = \frac{\partial}{\partial t} \int_\forall v_x \rho d\forall + \int_{SC} v_x \rho (\vec{V}\cdot d\vec{A})
$$

Não há forças de corpo na horizontal, pois a gravidade age na vertical. Assim:

$$
\sum F_{c, x} = 0
$$

Há duas forças de superfície, uma da pressão manométrica na entrada e a outra referente à reação dos parafusos. Assim:

$$
\sum F_{s, x} = p_e A_1 - F_p
$$

O sistema opera em regime. Portanto:

$$
\int_\forall v_x \rho d\forall = 0
$$

Os fluxos através da superfície de controle são dados por:

$$
\int_{SC} v_x \rho (\vec{V}\cdot d\vec{A}) = \int_{1} v_x \rho (\vec{V}\cdot d\vec{A}) + \int_{2} v_x \rho (\vec{V}\cdot d\vec{A}) = -\rho v_e^2A_1+\rho v_s^2 A_2
$$

Substituindo todos os termos:

$$
p_e A_1 - F_p = -\rho v_e^2A_1+\rho v_s^2 A_2
$$

Assim:

$$
F_p = p_e A_1+\rho v_e^2A_1-\rho v_s^2 A_2
$$

Os resultados seguem abaixo.

```python
# 2.d)

Fp = pe*A1 + rho*(ve**2)*A1 - rho*(vs**2)*A2

print('Fp = ', Fp, 'N')
```

```python
# 3)

%reset -f

# Dados

d = 2.0       # m
rho = 998.37  # kg/m^3
V1 = 3        # m/s
D1 = 0.05     # m
D2 = 0.07     # m
Q3 = 0.01     # m^3/s
g = 9.81      # m/s^2

# Medidas auxiliares

from numpy import pi

A1 = pi*(D1**2)/4
A2 = pi*(D2**2)/4
```

### Questão 3.a) - Solução

Considere o diagrama a seguir:

image.png

Após a definição do volume de controle conforme o diagrama, podemos aplicar o princípio da conservação da massa:

$$
0 = \frac{\partial}{\partial t} \int_\forall{\rho d\forall} + \int_{SC}{\rho \left(\vec{V} \cdot d\vec{A} \right)}
$$

O volume de controle não está em regime. A massa de fluido no volume de controle é dada por:

$$
m_{vc} = \int_\forall \rho d\forall =\rho h\left(\frac{\pi d^2}{4} \right)
$$

A variação desta massa é dada por:

$$
\frac{\partial m_{vc}}{\partial t} = \frac{\partial}{\partial t} \int_\forall \rho d\forall = \rho \left(\frac{\pi d^2}{4}\right)\frac{dh}{dt}
$$

O fluxo de massa pela superfície de controle é dado por:

$$
\int_{SC}{\rho \left(\vec{V} \cdot d\vec{A} \right)} = \int_{1}{\rho \left(\vec{V} \cdot d\vec{A} \right)} + \int_{2}{\rho \left(\vec{V} \cdot d\vec{A} \right)} + \int_{3}{\rho \left(\vec{V} \cdot d\vec{A} \right)} = -\rho v_1 \left( \frac{\pi D_1^2}{4} \right) + \rho v_2 \left( \frac{\pi D_2^2}{4} \right) -\rho Q_3
$$

Substituindo todos os termos:

$$
0 = \rho \left(\frac{\pi d^2}{4}\right)\frac{dh}{dt} -\rho v_1 \left( \frac{\pi D_1^2}{4} \right) + \rho v_2 \left( \frac{\pi D_2^2}{4} \right) -\rho Q_3
$$

Reorganizando:

$$
\frac{dh}{dt} = \frac{1}{d^2}\left( v_1 D_1^2 - v_2 D_2^2  + \frac{4 Q_3}{\pi} \right)
$$

### Questão 3.b) - Solução

Para a condição de equilíbrio, o nível do reservatório deixará de variar. Assim, pode-se assumir $dh/dt = 0$ ou aplicar o princípio da conservação da massa. De qualquer forma:

$$
v_1 A_1 - v_2 A_2  + Q_3=0
$$

Assim:

$$
v_2 = \frac{v_1 A_1 + Q_3}{A_2}
$$

Os resultados seguem abaixo.

```python
# 3.b)

V2 = (V1*A1+Q3)/A2

print('V2 = ', V2, 'm/s')
```

### Questão 3.c) - Solução

Para calcular o nível de equilíbrio, basta aplicar a equação de Bernoulli na condição de equilíbrio entre um ponto no nível do reservatório (A) e um na saída 2 (B).

No nível temos $p_A = P_{atm}$, $v_A = 0$, $z_A = h$. Na saída 2 temos: $p_B = P_{atm}$, $v_B = v_2$ e $z_B = 0$. Assim:

$$
\frac{p_A}{\rho} + \frac{v_A^2}{2} + g z_A = \frac{p_B}{\rho} + \frac{v_B^2}{2} + g z_B
$$

Substituindo os dados:

$$
\frac{P_{atm}}{\rho} + g h = \frac{P_{atm}}{\rho} + \frac{v_2^2}{2}
$$

Portanto:

$$
h = \frac{v_2^2}{2g}
$$

Os resultados seguem abaixo.

```python
h = V2**2/(2*g)

print('h = ', h, 'm')
```
