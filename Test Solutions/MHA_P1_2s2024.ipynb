### Questão 1 a)

Para resolver este exercício se utiliza o princípio do gradiente de pressão e dos vasos comunicantes.

Chama-se a pressão do ar dentro da cavidade de $P_{ar}$. Assim, a pressão na interface entre o benzeno e a água é dada por:

$$
P_1 = P_{ar} + \rho_b g h
$$

Portanto, a pressão na interface entre a água e o mercúrio é de:

$$
P_2 = P_1 + \rho_{H_2O}gh
$$

Esta pressão deve ser a mesma pressão encontrada no mesmo nível do outro lado do manômetro, já que o mercúrio está em uma configuração de vasos comunicantes. Assim, neste nível, a pressão $P_2$ é dada por:

$$
P_2 = P_{atm} + \rho_{Hg}g(2h)
$$

Assim:

$$
\begin{eqnarray}
P_{ar} + \rho_b gh & = & P_1 \\
& = & P_2 - \rho_{H_2O}gh \\
& = & P_{atm} + \rho_{Hg}g(2h) - \rho_{H_2O}gh
\end{eqnarray}
$$

Portanto:
$$
P_{ar} - P_{atm} = P_{ar}^{(mano)} = \rho_{Hg}g(2h) - \rho_{H_2O}gh - \rho_b g h
$$

Do enunciado:

- $h = 0,1\,\mathrm{m}$
- $\rho_b = 876\,\mathrm{kg/m^3}$
- $\rho_{H_2O} = 998\,\mathrm{kg/m^3}$
- $\rho_{Hg} = 13.600\,\mathrm{kg/m^3}$
- $g = 9,81\,\mathrm{m/s^2}$

Assim, encontra-se o resultado a seguir:

```python
rho_b = 876
rho_H2O = 998
rho_Hg = 13600
g = 9.81
h = 0.1

P_ar_mano = rho_Hg*g*(2*h) - rho_H2O*g*h - rho_b*g*h

print(f'P_ar (mano) = {P_ar_mano:.2f} Pa')
```

### Questão 1 b)

Quando a abertura for feita no tanque, a pressão do ar será igual à pressão atmosférica. Como no caso anterior a pressão manométrica do ar era positiva, quando esta pressão se igualar à atmosférica, um pequeno volume $\delta V$ do mercúrio que está na parte de menor diâmetro do manômetro se deslocará para a parte de maior diâmetro.

Isso se traduz em uma redução do nível do lado direito. Como o diâmetro deste lado é $d$, a variação de nível será de:

$$
\delta H = \frac{4\delta V}{\pi d^2}
$$

Do lado esquerdo, o nível do mercúrio aumentará, pela mesma quantidade de volume, uma altura $\delta h$ dada por:

$$
\delta h = \frac{4\delta V}{\pi D^2}
$$

E para os cálculos a seguir, é importante observar a relação:

$$
\delta h = \delta H\left(\frac{d^2}{D^2}\right)
$$

Pode-se calcular, de ambos os lados, a pressão no fundo do reservatório. Do lado esquerdo ela será dada por:

$$
P_f = P_{atm} + \rho_b g h + \rho_{H_2O} g h + \rho_{Hg}g(h+\delta h)
$$

Já do lado mais estreito, a pressão no fundo é dada por:

$$
P_f = P_{atm} + \rho_{Hg} g (3h-\delta H)
$$

Pelo princípio dos vasos comunicantes, as pressões no fundo do vaso são iguais. Portanto:

$$
P_{atm} + \rho_b g h + \rho_{H_2O} g h + \rho_{Hg}g(h+\delta h) = P_{atm} + \rho_{Hg} g (3h-\delta H)
$$

Aplicando as simplificações cabíveis e reorganizando:

$$
\rho_b g h + \rho_{H_2O} g h = \rho_{Hg} g (2h-\delta H-\delta h)
$$

Aplicando a relação entre $\delta h$ e $\delta H$ observada anteriormente e reorganizando mais termos:

$$
2\rho_{Hg} g h - \rho_b g h - \rho_{H_2O} g h = \rho_{Hg} g \delta H \left(1+\frac{d^2}{D^2}\right)
$$

Assim:

$$
\delta H = \frac{h\left(2\rho_{Hg}  - \rho_b  - \rho_{H_2O} \right)}{\displaystyle\rho_{Hg} \left(1+\frac{d^2}{D^2}\right)} = 0,1844\,\mathrm{m}
$$

Portanto o novo nível será de:

$$
0,3\,\mathrm{m} - \delta H = 0,1156\,\mathrm{m}
$$



```python
d = 0.025  # m
D = 0.25   # m

dH = h*(2*rho_Hg - rho_b - rho_H2O)/(rho_Hg*(1+(d**2/D**2)))

print(f'dH = {dH:.4f} m')

N = 0.3 - dH
print(f'Novo nível = {N:.4f}')
```

### Questão 2 a)

A força realizada pelo concreto no nível mostrado é dada por:

$$
F_{eq} = \rho g h_c A
$$

Neste caso, $h_c$ é a profundidade do centroide de área do triângulo, que é um terço do nível de $0,25\,\mathrm{m}$. A densidade do concreto é de $\rho = 2.400 \,\mathrm{kg/m^3}$, a aceleração da gravidade é de $9,81\,\mathrm{m/s^2}$ e a área do triângulo pode ser calculada a partir de sua base, que por sua vez pode ser calculada por semelhança de triângulos, dado que o triângulo isósceles de mesmo ângulo interno possui altura de $0,4\,\mathrm{m}$ e base de $0,3\,\mathrm{m}$.

Os resultados são calculados a seguir:

```python
%reset -f

rho = 2400  # kg/m^3
g = 9.81    # m/s²
h = 0.25    # m
hc = h/3
b = 0.3*0.25/0.4
A = b*h/2

F_eq = rho*g*hc*A

print(f'F_eq = {F_eq:.2f} N')
```

### Questão 2 b)

Para calcular a posição do centro de pressão em $x$, pode-se aplicar a equação:

$$
x' = x_c + \frac{I_{\hat{x}\hat{y}}}{y_c A}
$$

Como o triângulo é isósceles, ele possui um eixo de simetria vertical. Definindo a origem do sistema de coordenadas de forma que o eixo $y$ passe por este eixo de simetria, tem-se $x_c = 0$ e $I_{\hat{x}\hat{y}}=0$, portanto $x' = x_c = 0$.

Para calcular a posição do centro de pressão em $y$ é necessário utilizar:
$$
y' = y_c + \frac{I_{\hat{x}\hat{x}}}{y_c A}
$$

Se o triângulo isósceles possui base $b$ e altura $h$, então o momento de inércia de área $I_{\hat{x}\hat{x}}$ é dado por:

$$
I_{\hat{x}\hat{x}} = \frac{b^3h}{36}
$$

Os resultados se encontram a seguir:

```python
Ixx = (b**3)*h/36

yp = hc + Ixx/(hc*A)

print(f"y' = {yp:.4f} m")
```
