### Questão 1) - Solução

Para resolver este problema, basta aplicar a equação de Bernoulli na linha central do bocal. Nesta linha, que é horizontal, a elevação na entrada e na saída é praticamente a mesma. Assim, $z_i = z_o = 0$, por conveniência. As pressões na entrada ($p_i$) e na saída ($p_o$) são conhecidas e as velocidades se relacionam com a vazão volumétrica através de:

$$
Q = v_i A_i = v_o A_o
$$

Assim:

$$
v_i = v_o \left(\frac{A_o}{A_i}\right)
$$

Aplicando na equação de Bernoulli:

$$
\frac{p_i}{\rho} + \frac{v_i^2}{2} + g z_i = \frac{p_o}{\rho} + \frac{v_o^2}{2} + g z_o
$$

Substituindo os dados:
$$
\frac{p_i}{\rho} + \frac{v_o^2}{2}\left(\frac{A_o}{A_i}\right)^2 =  \frac{v_o^2}{2}
$$

Isolando $v_o$:
$$
v_o = \sqrt{\frac{2 p_i}{\displaystyle \rho\left(1-\left(\frac{A_o}{A_i}\right)^2\right)}}
$$

Finalmente:

$$
Q = v_o A_o
$$

Os resultados seguem:


```python
from numpy import pi, sqrt

Di = 3*0.0254  # m
Do = 1*0.0254  # m

Pi = 100*6894.76  # Pa

rho = 998.37  # kg/m^3

Ai = pi*(Di**2)/4
Ao = pi*(Do**2)/4

vo = sqrt(2*(Pi/rho)/(1-(Ao**2/Ai**2)))

print('v_o = ', vo, 'm/s')
print('Q = ', 1000*vo*Ao, 'L/s')
```

### Questão 2) - Solução

Para solucionar esta questão, basta aplicar a equação de Bernoulli entre os pontos A e B.

Para o ponto A, a pressão manométrica é dada por:

$$
p_a = \rho g . 0,165\,m
$$

A velocidade é dada por:

$$
v_a = \frac{Q}{A_1} = \frac{4 Q}{\pi (0,1\,m)^2}
$$

A elevação pode ser adotada como $z_a = 0$ por conveniência.

Para o ponto $B$, a pressão $p_b$ é o que se deseja descobrir. A velocidade $v_b$ é nula, pois se trata de um ponto de estagnação, e a elevação é de $z_b = 0,4\, m$

A Equação de Bernoulli entre estes pontos é dada por:

$$
\frac{p_a}{\rho} + \frac{v_a^2}{2} + g z_a = \frac{p_b}{\rho} + \frac{v_b^2}{2} + g z_b
$$

Isolando $p_b$:

$$
p_b = \rho \left(\frac{p_a}{\rho} + \frac{v_a^2}{2} + g z_a - \frac{v_b^2}{2} - g z_b\right)
$$

Conhecendo $p_b$, calcula-se $h$ através do princípio dos vasos comunicantes:

$$
h = \frac{p_b}{\rho g}
$$

Os resultados seguem.

```python
%reset -f

from numpy import pi

Q = 0.02      # m^3/s
rho = 998.37  # kg/m^3
g = 9.81      # m/s^2

h1 = 0.165  # m
D1 = 0.1    # m
D2 = 0.2    # m
H  = 0.4    # m

va = Q/(pi*(D1**2)/4)
za = 0
pa = rho*g*h1

vb = 0
zb = H

pb = rho*(pa/rho + va**2/2 + g*za - vb**2/2 - g*zb)

h = pb/(rho*g)

print('h = ', h*1000, 'mm')
```

### Questão 3) - Solução

A comporta em questão possui um dos lados submetido à pressão ambiente, a mesma pressão que é exercida sobre o nível do reservatório. Por isso pode-se utilizar as fórmulas para este caso específico.

A força exercida sobre a comporta é dada por:

$$
F_{q_{Eq}} = \rho g y_c \sin \theta A
$$

com $\theta = 90^\circ$ e todos os dados fornecidos.

A comporta é simétrica, portanto o centro de pressão na direção $x$ ficará sobre o centro geométrico da figura. Já na direção $y$, temos a relação:

$$
y' = y_c + \frac{I_{\hat{x}\hat{x}}}{y_c A}
$$

Neste caso, $y_c$ é a distância do centróide de área da comporta do nível, na direção do eixo $y$, que neste caso é vertical e positivo para baixo. Assim, $y_c = 2,5\,m$. O momento de inércia de área é dado por:

$$
I_{\hat{x}\hat{x}} = \frac{\pi r^4}{4}
$$

E a área é dada por $A = \pi r^2$.

Os resultados seguem:

```python
%reset -f

from numpy import pi

rho = 932  # kg/m^3
g = 9.81  # m/s^2

yc = 2.5  # m
r = 0.75  # m

A = pi*(r**2)

Feq = rho*g*yc*A

Ixx = pi*(r**4)/4

yp = yc + Ixx/(yc*A)

print('Fq_eq = ', Feq, 'N')
print("y' = ", yp, 'm')
print("y' - yc = ", yp-yc, 'm')
```
